```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("Hackathon_Working_Data.csv")
data.head()# Display the first few rows of the dataframe
```

```python
# Get summary statistics of numerical columns
print(data.describe())
```

```python
data.info()
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
data.select_dtypes(include=np.number)
```

```python
# Explore the distribution of categorical variables
for col in ['MONTH', 'STORECODE', 'GRP', 'SGRP', 'SSGRP', 'CMP', 'MBRD', 'BRD']:
    print(data[col].value_counts())
```

```python
# Visualize the distribution of BILL_AMT using a histogram
import matplotlib.pyplot as plt
plt.hist(data['BILL_AMT'])
plt.xlabel('Bill Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Bill Amount')
plt.show()
```

```python

# Explore the relationship between BILL_AMT and QTY using a scatter plot
plt.scatter(data['BILL_AMT'], data['QTY'])
plt.xlabel('Bill Amount')
plt.ylabel('Quantity')
plt.title('Relationship between Bill Amount and Quantity')
plt.show()
```
